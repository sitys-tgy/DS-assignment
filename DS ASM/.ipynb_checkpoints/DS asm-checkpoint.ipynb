### Data cleaning

```python
data=train_data
data.shape
```

```python
data['City_Code'] = data['City_Code'].str.split("_",expand=True)[1].astype('int64')

col_with_nan = ['Gender', 'Drafteed_University', 'Education_Level','Major_Discipline','Experience_Year','Agency_Size','Agency_Type','Pre_Employment_Duration']

for i in col_with_nan:
    data[i] = data[i].fillna(data[i].mode()[0])
```

```python
data = data[data.Gender != 'Other']
```

```python
data.info()
```

```python
le = LabelEncoder()

data['Gender']= le.fit_transform(data['Gender'])
data['Relevant_Experience']= le.fit_transform(data['Relevant_Experience'])
data['Drafteed_University']= le.fit_transform(data['Drafteed_University'])
data['Education_Level']= le.fit_transform(data['Education_Level'])
data['Major_Discipline']= le.fit_transform(data['Major_Discipline'])
data['Experience_Year']= le.fit_transform(data['Experience_Year'])
data['Agency_Size']= le.fit_transform(data['Agency_Size'])
data['Agency_Type']= le.fit_transform(data['Agency_Type'])
data['Pre_Employment_Duration']= le.fit_transform(data['Pre_Employment_Duration'])
```

```python
data.head()
```

```python
plt.figure(figsize=(12,10))
sns.heatmap(data.corr(), annot = True, color='b')
```

```python
X = data.drop(columns='target')
Y = data['target']

over = SMOTE(random_state=42)
X_smote, Y_smote = over.fit_resample(X,Y)
```

```python
Y_smote.value_counts()
```

### Data splitting

```python
X_train,X_test,Y_train,Y_test = train_test_split(X_smote,Y_smote,shuffle=True,random_state=42)
```
